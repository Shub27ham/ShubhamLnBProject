```python
### Q8
 
def spiralform(matrix):  
    ''''' k - beginning of the row 
        m - last row index 
        l - beginning of the column 
        n - last column index '''  
    k = 0  
    m = len(matrix)  
    l = 0  
    n = len(matrix[0])  
  
    # Starting the loop with the condition to stoop the loop when k > m and l > n  
    while (k < m and l < n):  
    
        # Printing the first row out of the unvisited rows from the remaining columns  
        for i in range(l, n):  
            print(matrix[k][i], end = " ")  
    
        k = k + 1  
    
        # Printing the rightmost column out of the remaining columns for the remaining rows  
        for i in range(k, m):  
            print(matrix[i][n - 1], end = " ")  
    
        n = n - 1  
    
        # Printing the last row from the remaining rows for the remaining columns  
        if (k < m):  
            for i in range(n - 1, (l - 1), -1):  
                print(matrix[m - 1][i], end = " ")  
    
            m = m - 1  
    
        # Printing the leftmost column out of the remaining columns  
        if (l < n):  
            for i in range(m - 1, k - 1, -1):  
                print(matrix[i][l], end = " ")  
    
            l = l + 1  
    
  
# Calling the function and passing the matrix to it  
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
  
  
spiralform(matrix)  
```

```python
### Q9
M = 9
def puzzle(a):
	for i in range(M):
		for j in range(M):
			print(a[i][j],end = " ")
		print()
def solve(grid, row, col, num):
	for x in range(9):
		if grid[row][x] == num:
			return False
		    
	for x in range(9):
		if grid[x][col] == num:
			return False


	startRow = row - row % 3
	startCol = col - col % 3
	for i in range(3):
		for j in range(3):
			if grid[i + startRow][j + startCol] == num:
				return False
	return True

def Suduko(grid, row, col):

	if (row == M - 1 and col == M):
		return True
	if col == M:
		row += 1
		col = 0
	if grid[row][col] > 0:
		return Suduko(grid, row, col + 1)
	for num in range(1, M + 1, 1): 
	
		if solve(grid, row, col, num):
		
			grid[row][col] = num
			if Suduko(grid, row, col + 1):
				return True
		grid[row][col] = 0
	return False

'''0 means the cells where no value is assigned'''
grid = [[2, 5, 0, 0, 3, 0, 9, 0, 1],
        [0, 1, 0, 0, 0, 4, 0, 0, 0],
	[4, 0, 7, 0, 0, 0, 2, 0, 8],
	[0, 0, 5, 2, 0, 0, 0, 0, 0],
	[0, 0, 0, 0, 9, 8, 1, 0, 0],
	[0, 4, 0, 0, 0, 3, 0, 0, 0],
	[0, 0, 0, 3, 6, 0, 0, 7, 2],
	[0, 7, 0, 0, 0, 0, 0, 0, 3],
	[9, 0, 3, 0, 0, 0, 6, 0, 4]]

if (Suduko(grid, 0, 0)):
	puzzle(grid)
else:
	print("Solution does not exist:(")
```

```python
### Q7
def lcs_algo(S1, S2, m, n):
    L = [[0 for x in range(n+1)] for x in range(m+1)]

    # Building the mtrix in bottom-up way
    for i in range(m+1):
        for j in range(n+1):
            if i == 0 or j == 0:
                L[i][j] = 0
            elif S1[i-1] == S2[j-1]:
                L[i][j] = L[i-1][j-1] + 1
            else:
                L[i][j] = max(L[i-1][j], L[i][j-1])

    index = L[m][n]

    lcs_algo = [""] * (index+1)
    lcs_algo[index] = ""

    i = m
    j = n
    while i > 0 and j > 0:

        if S1[i-1] == S2[j-1]:
            lcs_algo[index-1] = S1[i-1]
            i -= 1
            j -= 1
            index -= 1

        elif L[i-1][j] > L[i][j-1]:
            i -= 1
        else:
            j -= 1
            
    # Printing the sub sequences
    print("S1 : " + S1 + "\nS2 : " + S2)
    print("LCS: " + "".join(lcs_algo))


S1 = "AGGTAB"
S2 = "GXTXAYB"
m = len(S1)
n = len(S2)
lcs_algo(S1, S2, m, n)
```

```python
### Q6
def maxSubArraySum(arr,size):
    
    max_till_now = arr[0]
    max_ending = 0
    
    for i in range(0, size):
        max_ending = max_ending + arr[i]
        if max_ending < 0:
            max_ending = 0
        
        
        elif (max_till_now < max_ending):
            max_till_now = max_ending
            
    return max_till_now

arr = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
print("Maximum Sub Array Sum Is" , maxSubArraySum(arr,len(arr)))
```
